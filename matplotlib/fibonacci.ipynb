```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from functools import lru_cache
```

```python
@np.vectorize
@lru_cache(maxsize=1000)
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
x = np.arange(0, 1000, dtype=np.float128)
y = fibonacci(x)

plt.plot(x, y)
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
y1 = y[1:-1]
y2 = y[2:]
div = y2 / y1
plt.plot(x[2:], div)
plt.grid(True)
plt.xscale('log')
plt.show()
```

```python
print(f'''\
avg: {round(np.average(div), 5)} 
std: {round(np.std(div), 5)}''')
```
